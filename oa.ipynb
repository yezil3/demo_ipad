```python
def getUniqueCharacter(a_string):
    list_char=[]
    for i in range(len(a_string)):
        if a_string[i] in list_char:
            list_char.remove(a_string[i])
            continue
        else:
            list_char.append(a_string[i])
        i+=1
    if len(list_char)==0:
        return -1
    else:
        return a_string.index(list_char[0])
```

```python
def firstUniqChar(s):
    mp = {}

    for a in s:
        mp[a] = mp.get(a, 0) + 1

    for i in range(len(s)):
        if mp[s[i]] == 1:
            return i

    return -1
```

```python
def merge_v1(intervals):
    intervals.sort(key =lambda x: x[0])
    merged =[]
    for i in intervals:
        # if the list of merged intervals is empty 
        # or if the current interval does not overlap with the previous,
        # simply append it.
        if not merged or merged[-1][-1] < i[0]:
            merged.append(i)
        # otherwise, there is overlap,
        #so we merge the current and previous intervals.
        else:
            merged[-1][-1] = max(merged[-1][-1], i[-1])
    return merged
```

```python
list_intervals=[[7,7],[2,3],[6,11],[1,2],[2,3],[4,5],[5,7]]
merge_v1(list_intervals)
```

```python
def merge_v2(intervals):
    intervals.sort()
    merged = []
    for i in range(len(intervals)):
        if merged == []:
            merged.append(intervals[i])
        else:
            previous_end = merged[-1][1]
            current_start = intervals[i][0]
            current_end = intervals[i][1]
            if previous_end >= current_start: # overlap
                merged[-1][1] = max(previous_end,current_end)
            else:
                merged.append(intervals[i])
    return merged
                    
```

```python
list_intervals=[[7,7],[2,3],[6,11],[1,2],[2,3],[4,5],[5,7]]
merge_v2(list_intervals)
```

```python
def merge_v3(intervals):
    intervals = sorted(intervals, key=lambda x: x [0])

    ans = []

    for interval in intervals:
        if not ans or ans[-1][1] < interval[0]:
            ans.append(interval)
        else:
            ans[-1][1] = max(ans[-1][1], interval[1])
    
    return ans
```

```python
list_intervals=[[7,7],[2,3],[6,11],[1,2],[2,3],[4,5],[5,7]]
merge_v3(list_intervals)
```

```python
def maximumFinal(arr):
    # Constraint 1: [1, ..., ]
    # Constraint 2: arr[i]-arr[i-1]<=1
    # given an array, return the maximum value. In the sorted array, neighbor distance >1 reduce
    # [2,4,6,8]->[1,2,3,4], max(max_value)=length(arr)
    arr.sort()
    arr[0]=1
    for i in range(1,len(arr)):
        if arr[i]-arr[i-1]>1:
            arr[i]=arr[i-1]+1
    # print(arr)        
    max_value=arr[-1]
    return max_value

# if __name__ == '__main__':-
```

```python
def maximumElementAfterDecrementingAndRearranging_v1(arr):
    arr.sort()
    res = 1
    for i in range(1,len(arr)):
        if (arr[i] > res):
            res += 1
    return res
```

```python
def getDistinctPairs_v1(stocksProfit, target):
    # num_pairs=0
    distinct_pair_dic=set()
    for i in range(0,len(stocksProfit)-1):
        for j in range(0,len(stocksProfit)-1):
            if stocksProfit[i]+stocksProfit[j]==target:
                distinct_pair_dic.add(tuple(sorted((stocksProfit[i],stocksProfit[j]))))
    print(distinct_pair_dic)
                
    num_pairs=len(distinct_pair_dic)
    return num_pairs
```

```python
def uniqueTwoSum_v2(nums, target):
  ans, comp = set(), set()
  for n in nums:
    c = target-n
    if c in comp:
      res = (n, c) if n > c else (c, n)
      if res not in ans:
        ans.add(res)
    comp.add(n)
  return len(ans)
```

```python
stocksProfit=[46,1,3,46,1,3,9,45,2]
target=47
uniqueTwoSum_v2(stocksProfit,target)
```
